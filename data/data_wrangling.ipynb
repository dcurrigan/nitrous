```python
metro_df = metro_df[metro_df['name'] != 'METRO HEALTH SERVICE BOARD']
metro_df
```

```python
metro_df[['cylinder_c', 'cylinder_d', 'cylinder_e', 'cylinder_f8', 'cylinder_f9', 
          'cylinder_g', 'total']] = ["", "", "", "", "", "", ""]
metro_df
```

```python
cylinder_list = ['cylinder_c', 'cylinder_d', 'cylinder_e', 'cylinder_f8', 'cylinder_f9', 'cylinder_g', 'total']

for index, row in metro_df.iterrows():
    cylinder = row['cylinder']
    for item in cylinder_list:
        if item == cylinder:
            metro_df.loc[index, item] = row['units']
        else:
            metro_df.loc[index, item] = 0
    
metro_df
```

```python
merged_data = pd.DataFrame(columns={'date', 'name', 'cylinder_c', 'cylinder_d', 'cylinder_e', 'cylinder_f8', 'cylinder_f9', 'cylinder_g'})

purchase_dates = metro_df['date'].unique()
hospital_list = metro_df['name'].unique()
cylinder_list = ['cylinder_c', 'cylinder_d', 'cylinder_e', 'cylinder_f8', 'cylinder_f9', 'cylinder_g']

for date in purchase_dates:
    for hospital in hospital_list:
        # did the current hospital have a purchase on this date         
        current_hosp = metro_df[metro_df['name'] == hospital]
        matched_purchase_dates = current_hosp[current_hosp['date'] == date]
        
        # if so, iterate through the purchases on this date and add to merged df        
        if (len(matched_purchase_dates) != 0):
                      
            # temp df to append to the final merged data
            new_row = pd.DataFrame(columns={'date', 'name', 'cylinder_c', 'cylinder_d', 'cylinder_e', 'cylinder_f8', 'cylinder_f9', 'cylinder_g'})
            new_row.loc[0, 'date'] = date
            new_row.loc[0, 'name'] = hospital
            
            # add each cylinder purchase to the new row and append
            for index, row in matched_purchase_dates.iterrows():
                for item in cylinder_list:
                    
                    if row[item] != 0:
                        new_row[item][0] = row[item]

                    elif (row[item] == 0) & (pd.isnull(new_row[item][0]) == True):
                        new_row.loc[0, item] = 0
                    
            merged_data = merged_data.append(new_row).reset_index(drop=True)
                        
merged_data = merged_data[['date', 'name', 'cylinder_c', 'cylinder_d', 'cylinder_e', 'cylinder_f8', 'cylinder_f9', 'cylinder_g']]
merged_data
                        
```

```python
# Add total column
final_df = merged_data.copy()
cylinder_list = ['cylinder_c', 'cylinder_d', 'cylinder_e', 'cylinder_f8', 'cylinder_f9', 'cylinder_g']

# Helper function to return total 
def total_calculator(row):
    total = 0
    amount = 0
    
    for cylinder in cylinder_list:
        if cylinder == 'cylinder_c':
            amount = row[cylinder] * 1.75
        elif cylinder == 'cylinder_d':
            amount = row[cylinder] * 6.6
        elif cylinder == 'cylinder_e':
            amount = row[cylinder] * 16.8
        elif cylinder == 'cylinder_f8':
            amount = row[cylinder] * 233
        elif cylinder == 'cylinder_f9':
            amount = row[cylinder] * 223
        elif cylinder == 'cylinder_g':
            amount = row[cylinder] * 36.6
            
        total += amount
    
    return(float(total))

for row in range (0, len(final_df)):
    total = total_calculator(final_df.iloc[[row]])
    final_df.loc[row, 'total'] = total
    
final_df
```
